```python
import pandas as pd

df=pd.read_csv("olympic-medals.csv",index_col=0)
df
```

```python
medallas=df.head(10)
medallas
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.bar(medallas.index,medallas["Gold"])
ax.set_xticklabels(medallas.index,rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.bar(medallas.index,medallas["Gold"])
ax.bar(medallas.index,medallas["Silver"],bottom=medallas["Gold"])
ax.set_xticklabels(medallas.index,rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.bar(medallas.index,medallas["Gold"])
ax.bar(medallas.index,medallas["Silver"],bottom=medallas["Gold"])
ax.bar(medallas.index,medallas["Bronze"],bottom=medallas["Gold"] + medallas["Silver"])
ax.set_xticklabels(medallas.index,rotation=90)
ax.set_ylabel("Numero de medallas")
ax.legend(["Gold","Silver","Bronze"])
plt.show()
```

```python
##CreaciÃ³n histograma
fig,ax=plt.subplots()
ax.hist(df["Gold"])
plt.show()
```

```python
fig,ax=plt.subplots()
ax.hist(df["Gold"])
ax.hist(df["Silver"])
ax.set_xlabel("Numero de medallas")
ax.set_ylabel("Numero de Observaciones")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.hist(df["Gold"],label="Gold")
ax.hist(df["Silver"],label="Silver")
ax.set_xlabel("Numero de medallas")
ax.set_ylabel("Numero de Observaciones")
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots()
ax.hist(df["Gold"],label="Gold",bins=[0,5,10,15,20,25,30,35,40],histtype="step")
ax.hist(df["Silver"],label="Silver",bins=[0,5,10,15,20,25,30,35,40],histtype="step")
ax.set_xlabel("Numero de medallas")
ax.set_ylabel("Numero de Observaciones")
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots()
ax.boxplot([df["Gold"],df["Silver"]])
ax.set_xticklabels(["Oro","Plata"])
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(df["Gold"],df["Silver"])
ax.set_xlabel("Medallas de oro")
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(df["Gold"],df["Silver"],color="red",label="Plata")
ax.scatter(df["Gold"],df["Bronze"],color="blue",label="Bronce")
ax.set_xlabel("Medallas de oro")
ax.set_ylabel("Numero de medallas")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(df["Gold"],df["Silver"],color=df.index)
ax.set_xlabel("Medallas de oro")
ax.set_ylabel("Medallas de plata")
plt.show()
```
