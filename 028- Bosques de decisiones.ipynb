```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
pima= pd.read_csv("diabetes.csv", header=1)
pima.head()
```

```python
df= pima.iloc[1:,:]
df
```

```python
col_names = ["pregnant","glucose","bp","skin","insulin","bmi","pedigree","age","label"]
pima= pd.read_csv("diabetes.csv",header=1, names= col_names)
pima.head()
```

```python
feature_cols = ["pregnant","glucose","bp","insulin","bmi","pedigree","age"]
X= df[feature_cols]
y=df.label
```

```python
print(df.columns.tolist())
```

**Bosques Aleatorios (Random Forest)**

```python
from sklearn.ensemble import RandomForestClassifier
##Crear los grupos de entrenamiento y prueba
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,random_state=1)
```

```python
rf= RandomForestClassifier(n_estimators=100, random_state=1)
rf.fit(X_train, y_train)
```

```python
##Score F1 para el grupo de entrenamiento
rf.score(X_train,y_train)
```

```python
##Score F1 para el grupo de testing
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

y_pred = rf.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_feature_importance(df, model):
    features = df.columns
    importances = model.feature_importances_
    
    # Sort features by importance
    indices = np.argsort(importances)
    
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(features)), importances[indices], align="center")
    plt.yticks(range(len(features)), features[indices])
    plt.xlabel("Importancia de las variables")
    plt.ylabel("Características")
    plt.title("Importancia de características - Random Forest")
    plt.show()

# Uso:
# plot_feature_importance(df, rf)
```
