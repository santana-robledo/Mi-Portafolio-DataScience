```python
import pandas as pd
df=pd.read_csv("StudentsPerformance.csv")
df
```

```python
import seaborn as sns
for style in ["dark","whitegrid","darkgrid","ticks"]:
    sns.set_style(style)
    sns.displot(df["reading score"])
```

```python
#Remover ejes
sns.set_style("ticks")
sns.displot(df["reading score"],kde=True)
sns.despine(left=True,bottom=True)
```

```python
#CombinaciÃ³n de parÃ¡metros con Matplotlib
import matplotlib.pyplot as plt

fig,ax=plt.subplots()
sns.distplot(df["writing score"],ax=ax)
ax.set(xlabel="Writing Score", ylabel="Frecuencia relativa",title="Writing Score")
ax.set_title("Writing Score")
```

```python
#Truncamiento de escala de visualizaciÃ³n
fig,ax=plt.subplots()
sns.distplot(df["writing score"],ax=ax)
ax.set(xlabel="Writing Score", ylabel="Frecuencia relativa",xlim=(0,90),title="Writing Score")
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=1,ncols=2,sharey=True)
sns.distplot(df["writing score"],ax=ax0) #Histograma para todas las observaciones
sns.distplot(df.query('gender == "male"')["writing score"], ax=ax1)#Histograma para los hombres
ax0.set(xlabel="Puntaje general de escritura",ylabel="Frecuencia")
ax1.set(xlabel="Puntaje de escritura para hombres",ylabel="Frecuencia")
ax1.axvline(x=75,label="Objetivo",linestyle="--")
ax1.legend()
```

```python
#AsignaciÃ³n de colores al histograma
paleta=["deep","muted","pastel","bright","dark","colorblind"]
for p in paleta:
    sns.set_palette(p)
    sns.displot(df["reading score"],kde=True)
```
