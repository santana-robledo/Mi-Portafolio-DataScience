```python
import pandas as pd
df=pd.read_csv("StudentsPerformance.csv")
df
```

```python
import seaborn as sns
sns.stripplot(data=df, y="writing score", x="math score", jitter=True)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
sns.stripplot(data=df, y="writing score", x="math score", jitter=True)
ax.axes.xaxis.set_ticks([])
plt.show
```

```python
fig,ax=plt.subplots()
sns.stripplot(data=df, y="writing score", x="race/ethnicity", jitter=True)
plt.show
```

```python
#Diagrama de enjambre
sns.swarmplot(data=df,y="writing score",x="race/ethnicity")
```

```python
#Diagrama de caja
sns.boxplot(data=df,y="writing score",x="race/ethnicity",palette="Set1")
```

```python
#Grafico de violin
sns.violinplot(data=df,y="writing score",x="race/ethnicity",palette="Set1")
```

```python
#creaciÃ³n de lvplot
sns.catplot(data=df,y="writing score",x="race/ethnicity",palette="Set1",kind="boxen")
```

```python
#creaciÃ³n de barplot
sns.barplot(data=df,y="writing score",x="race/ethnicity",palette="Set1")
```

```python
#creaciÃ³n de barplot
sns.barplot(data=df,y="writing score",x="race/ethnicity",palette="Set1",hue="lunch")
```

```python
#creaciÃ³n de pointplot
sns.pointplot(data=df,y="writing score",x="race/ethnicity",palette="Set1",hue="lunch")
```

```python
#creaciÃ³n de countplot
sns.countplot(data=df,y="race/ethnicity",palette="Set1",hue="lunch")
```

```python
#creaciÃ³n de grafico de regresiÃ³n
sns.regplot(data=df,y="writing score",x="reading score",marker="+")
```

```python
#creaciÃ³n de residuos
sns.residplot(data=df,y="writing score",x="reading score")
```

```python
import numpy as np
#creaciÃ³n de grafico de regresiÃ³n y media por valor dex
sns.regplot(data=df,y="writing score",x="reading score",x_estimator=np.mean)
```

```python
df_withoutString = df.drop(['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'], axis=1)
sns.heatmap(df_withoutString.corr())
```

```python
sns.heatmap(df_withoutString.corr(),annot=True)
```

```python
#CreaciÃ³n de tabla cruzada
pd.crosstab(df["gender"],df["race/ethnicity"],values=df["math score"],aggfunc="mean")
```

```python
pivot_table = pd.pivot_table(df, 
                              values="math score", 
                              index="gender", 
                              columns="race/ethnicity", 
                              aggfunc="mean")

# Crear el heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu")
plt.title("Promedio de Math Score por GÃ©nero y Grupo Ã‰tnico")
plt.show()
```
