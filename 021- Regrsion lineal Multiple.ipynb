```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.linear_model import LinearRegression
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
url="http://archive.ics.uci.edu/ml/machine-learning-databases/communities/communities.data"
crime=pd.read_csv(url,header=None,na_values=["?"])
crime.head()
```

```python
crime[127].describe()
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(crime.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
#Remover caracteristicas categoricas
crime.drop([0,1,2,3,4],axis=1,inplace=True)
crime
```

```python
#Remover renglones con valores faltante
crime.dropna(inplace=True)
crime
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(crime.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
crime.shape
```

```python
x=crime.drop(127,axis=1)
y=crime[127]
x
```

```python
y
```

```python
##DivisiÃ³n en grupos de entrenamiento y de prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
##Construcciond e un modelo de regresion linear multiple
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
##Impresion de coeficientes de regresiÃ³n
print("Intercepto: ",linreg.intercept_)
print("Coeficientes: ",linreg.coef_)
```

```python
##Prediccion
y_pred=linreg.predict(x_test)
y_pred
```

```python
#Impresion de indicadores de la bondad de ajuste
print("Valor de r cuadrada: ",r2_score(y_test,y_pred))
print("Error absoluto medio: ",metrics.mean_absolute_error(y_test,y_pred))
print("Error cuadratico medio ",metrics.mean_squared_error(y_test,y_pred))
print("Raiz del error cuadratico medio: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#Alfa=0 (Equivalente a regresiÃ³n linea)
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn import metrics
from sklearn.metrics import r2_score

# Crear un pipeline: primero escalar, luego ajustar el modelo
ridgereg = make_pipeline(StandardScaler(), Ridge(alpha=0))
ridgereg.fit(x_train, y_train)
y_pred = ridgereg.predict(x_test)

# EvaluaciÃ³n
print("Valor de R^2: ", r2_score(y_test, y_pred))
print("Error absoluto medio: ", metrics.mean_absolute_error(y_test, y_pred))
print("Error cuadrÃ¡tico medio: ", metrics.mean_squared_error(y_test, y_pred))
print("RaÃ­z del error cuadrÃ¡tico medio: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

<h1 style="color:blue;"><b>RegresiÃ³n Ridge</b></h1>

```python
#Alfa=0.1 (Equivalente a regresiÃ³n linea)
# Crear un pipeline: primero escalar, luego ajustar el modelo
ridgereg = make_pipeline(StandardScaler(), Ridge(alpha=0.1))
ridgereg.fit(x_train, y_train)
y_pred = ridgereg.predict(x_test)

# EvaluaciÃ³n
print("Valor de R^2: ", r2_score(y_test, y_pred))
print("Error absoluto medio: ", metrics.mean_absolute_error(y_test, y_pred))
print("Error cuadrÃ¡tico medio: ", metrics.mean_squared_error(y_test, y_pred))
print("RaÃ­z del error cuadrÃ¡tico medio: ", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#ExaminaciÃ³n de los coeficioentes de la regresiÃ³n Ridge
# Acceder al modelo Ridge dentro del pipeline
modelo_ridge = ridgereg.named_steps['ridge']
print("Intercepto: ", modelo_ridge.intercept_)
print("Coeficientes: ", modelo_ridge.coef_)
```

```python
#Rango de prueba para valor Alpha
alpha_range = 10. **np.arange(-3, 2)
alpha_range
```

```python
from sklearn.linear_model import RidgeCV
ridgecv_pipeline = make_pipeline(
    StandardScaler(),
    RidgeCV(alphas=alpha_range, scoring="neg_mean_squared_error")
)

ridgecv_pipeline.fit(x_train, y_train)

# Acceder al mejor valor de alpha:
best_alpha = ridgecv_pipeline.named_steps["ridgecv"].alpha_
print("Mejor valor de alpha:", best_alpha)
```

```python
#Prediccion mediante el mejor valor de Alpha
y_pred = ridgecv_pipeline.predict(x_test)
print("Valor de r cuadrada: ",r2_score(y_test,y_pred))
print("Error absoluto medio: ",metrics.mean_absolute_error(y_test,y_pred))
print("Error cuadratico medio ",metrics.mean_squared_error(y_test,y_pred))
print("Raiz del error cuadratico medio: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
modelo_ridge = ridgereg.named_steps['ridge']
print("Intercepto: ", modelo_ridge.intercept_)
print("Coeficientes: ", modelo_ridge.coef_)
```

<h1 style="color:blue;"><b>RegresiÃ³n Lasso</b></h1>

```python
#Prueba con Alpha=0.001
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lassoreg = make_pipeline(StandardScaler(), Lasso(alpha=0.001))
lassoreg.fit(x_train, y_train)

# Acceder a los coeficientes del modelo Lasso dentro del pipeline
modelo_lasso = lassoreg.named_steps['lasso']
print("Intercepto: ", modelo_lasso.intercept_)
print("Coeficientes: ", modelo_lasso.coef_)
```

```python
#Prueba con Alpha=0.001
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lassoreg = make_pipeline(StandardScaler(), Lasso(alpha=0.01))
lassoreg.fit(x_train, y_train)

# Acceder a los coeficientes del modelo Lasso dentro del pipeline
modelo_lasso = lassoreg.named_steps['lasso']
print("Intercepto: ", modelo_lasso.intercept_)
print("Coeficientes: ", modelo_lasso.coef_)
```

```python
#DeterminaciÃ³n del grado de bondad de ajuste con Alpha =0.01
y_pred = lassoreg.predict(x_test)
print("Valor de r cuadrada: ",r2_score(y_test,y_pred))
print("Error absoluto medio: ",metrics.mean_absolute_error(y_test,y_pred))
print("Error cuadratico medio ",metrics.mean_squared_error(y_test,y_pred))
print("Raiz del error cuadratico medio: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#SelecciÃ³n del valor optimo de Alpha para la regresiÃ³n lasso
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Crear pipeline con escalado y LassoCV
lassoregcv = make_pipeline(
    StandardScaler(),
    LassoCV(n_alphas=100, random_state=1)
)

lassoregcv.fit(x_train, y_train)

# Acceder al mejor alpha encontrado
modelo_lasso = lassoregcv.named_steps["lassocv"]
print("Alpha Ã³ptimo: ", modelo_lasso.alpha_)
```

```python
# Acceder a los coeficientes del modelo Lasso dentro del pipeline
modelo_lasso = lassoreg.named_steps['lasso']
print("Intercepto: ", modelo_lasso.intercept_)
print("Coeficientes: ", modelo_lasso.coef_)
```

```python
#Prediccion mediante regresiÃ³n de Lasso con un Alpha Ã³ptimo
y_pred = lassoregcv.predict(x_test)

#ImpresiÃ³n de indicadores de bondad de ajuste
print("Valor de r cuadrada: ",r2_score(y_test,y_pred))
print("Error absoluto medio: ",metrics.mean_absolute_error(y_test,y_pred))
print("Error cuadratico medio ",metrics.mean_squared_error(y_test,y_pred))
print("Raiz del error cuadratico medio: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
