```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig,ax =plt.subplots()
plt.show
```

```python
#Adicion de datos eje x
meses= pd.DataFrame(np.array(["Ene","Feb","Mar","Abr","May","Jun","Jul","Ago","Sep","Oct","Nov","Dic"]),columns=["Meses"])
meses
```

```python
#Adicion de datos al eje Y (Temperaturas promedio)
temperaturas=pd.DataFrame(np.array([20.3,22.5,18.6,21.2,23.3,23.6,26.4,28.5,25.1,23.3,22.1,21.9]),columns=["Temperaturas"])
temperaturas
```

```python
fig,ax =plt.subplots()
ax.plot(meses["Meses"],temperaturas["Temperaturas"])
plt.show
```

```python
temperaturas_x=pd.DataFrame(np.array([18.3,20.5,16.6,19.2,21.3,21.6,23.4,26.5,23.1,21.3,20.1,19.9]),columns=["Temperaturas Promedio"])
fig,ax =plt.subplots()
ax.plot(meses["Meses"],temperaturas["Temperaturas"])
ax.plot(meses["Meses"],temperaturas_x["Temperaturas Promedio"])
plt.show
```

```python
temperaturas_x=pd.DataFrame(np.array([18.3,20.5,16.6,19.2,21.3,21.6,23.4,26.5,23.1,21.3,20.1,19.9]),columns=["Temperaturas Promedio"])
fig,ax =plt.subplots()
ax.plot(meses["Meses"],temperaturas["Temperaturas"])
ax.plot(meses["Meses"],temperaturas_x["Temperaturas Promedio"])
ax.set_xlabel("Mes")
ax.set_ylabel("Grados C")
plt.show
```

```python
temperaturas_x=pd.DataFrame(np.array([18.3,20.5,16.6,19.2,21.3,21.6,23.4,26.5,23.1,21.3,20.1,19.9]),columns=["Temperaturas Promedio"])
fig,ax =plt.subplots()
ax.plot(meses["Meses"],temperaturas["Temperaturas"],color="r")
ax.plot(meses["Meses"],temperaturas_x["Temperaturas Promedio"],color="b")
ax.set_xlabel("Mes")
ax.set_ylabel("Grados C")
plt.show
```

```python
fig,ax =plt.subplots(2,1)
ax[0].plot(meses["Meses"],temperaturas["Temperaturas"],color="r")
ax[1].plot(meses["Meses"],temperaturas_x["Temperaturas Promedio"],color="g")
ax[1].set_xlabel("Mes")
ax[0].set_xlabel("Mes")
ax[0].set_ylabel("Grados C")
ax[1].set_ylabel("Temperatura Ciudad")
plt.show
```

```python
fig,ax =plt.subplots(2,1,sharey=True)
ax[0].plot(meses["Meses"],temperaturas["Temperaturas"],color="r")
ax[1].plot(meses["Meses"],temperaturas_x["Temperaturas Promedio"],color="g")
ax[1].set_xlabel("Mes")
ax[0].set_xlabel("Mes")
ax[0].set_ylabel("Grados C")
ax[1].set_ylabel("Temperatura Ciudad")
plt.show
```

```python
#Para subgraficos de mas de 1 renglon y mas de 1 columna
fig,ax =plt.subplots(3,2,sharey=True)
ax[0,0].plot(meses["Meses"],temperaturas["Temperaturas"],color="g")
ax[0,1].plot(meses["Meses"],temperaturas["Temperaturas"],color="r")
ax[0,0].set_ylabel("Temperatura Ciudad")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
        'Temperatura': [5, 7, 10, 15]}
df = pd.DataFrame(data)

# Crear un grfico de lneas
plt.plot(df['Mes'], df['Temperatura'], marker='o')
plt.title('Temperaturas promedio mensuales')
plt.xlabel('Mes')
plt.ylabel('Temperatura (C)')
plt.grid(True)
plt.show()
```
