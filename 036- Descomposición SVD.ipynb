```python
import os
import pandas as pd
```

```python
iris = pd.read_csv("iris.csv")
iris
```

```python
etiquetas = iris["variety"]
etiquetas
```

```python
from numpy import array
from sklearn.decomposition import TruncatedSVD
A= iris
#Remover columna no númerica
A= A.drop(["variety"], axis=1)
A
```

```python
from numpy import zeros
from numpy import diag
from scipy.linalg import svd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
U,s,VT= svd(A)
Sigma= zeros((A.shape[0],A.shape[1]))
Sigma[:A.shape[1],:A.shape[1]]=diag(s)
```

```python
print("Matriz U:")
print(U)
```

```python
print("Matriz Sigma:")
print(Sigma)
```

```python
print("Matriz VT:")
print(VT)
```

```python
#Para reconstruir la matriz original
B= U.dot(Sigma.dot(VT))
print(B)
```

```python
#Reduccion a dos dimensiones
n_elements=2
UReduced2= U[:,:n_elements]
print("U Reducida")
print(UReduced2)
```

```python
#Reduccion a dos dimensiones
SigmaReduced2= Sigma[:n_elements,:n_elements]
print("Sigma Reducida")
print(SigmaReduced2)
```

```python
#Reduccion a dos dimensiones
VTReduced2= VT[:n_elements,:]
print("VT Reducida")
print(VTReduced2)
```

```python
AReduced2= UReduced2.dot(SigmaReduced2.dot(VTReduced2))
print("Matriz A transformada:")
print(AReduced2)
```

```python
T2= UReduced2.dot(SigmaReduced2)
print("Matriz T (2 Dimensiones)")
print(T2)
```

```python
df = pd.DataFrame(T2,columns=["a","b"])
df.plot(kind="scatter",x="a",y="b")
plt.show
```

```python
df = pd.DataFrame(T2,columns=["a","b"])
x= df.iloc[:,0]
y= df.iloc[:,1]
x=x.to_numpy()
y=y.to_numpy()

fig,ax = plt.subplots()
ax.set_title("Mapa de observaciones")
ax.scatter(x,y)

for i,txt in enumerate(etiquetas):
    if txt=="Versicolor":
        ax.annotate(txt, (x[i],y[i]))
```

```python
#Reduccion a 3 dimensiones
n_elements=3
UReduced3= U[:,:n_elements]
print("U Reducida")
print(UReduced3)
```

```python
#Reduccion a dos dimensiones
SigmaReduced3= Sigma[:n_elements,:n_elements]
print("Sigma Reducida")
print(SigmaReduced3)
```
