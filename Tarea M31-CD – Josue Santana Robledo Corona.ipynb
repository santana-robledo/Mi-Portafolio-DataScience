```python
dataset = np.array([10,15,12,8,17,14,20,8,12,16,14])
```

```python
valores, conteos = np.unique(dataset, return_counts=True)
moda = valores[np.argmax(conteos)]
media= np.mean(dataset)
mediana= np.median(dataset)
maximo= np.max(dataset)
minimo= np.min(dataset)
varianza= np.var(dataset)
desviacion_tipica= np.std(dataset)
print(f"Media: {media:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Maximo: {maximo:.2f}")
print(f"Minimo: {minimo:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviacion típica: {desviacion_tipica:.2f}")
```
