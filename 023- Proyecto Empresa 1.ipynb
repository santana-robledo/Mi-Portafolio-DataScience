```python
import pandas as pd

# Carga de archivos
df_product = pd.read_excel('DIM_PRODUCT.xlsx')
df_segment = pd.read_excel('DIM_SEGMENT.xlsx')
df_calendar = pd.read_excel('DIM_CALENDAR.xlsx')
df_sales = pd.read_csv('FACT_SALES.csv')
```

```python
# Revisión básica
for name, df in [('SALES', df_sales), ('CALENDAR', df_calendar), 
                 ('SEGMENT', df_segment), ('PRODUCT', df_product)]:
    print(f"\n--- {name} ---")
    print(df.head())
    print(df.info())
    print(df.describe(include='all'))
```

```python
# Eliminación de duplicados
df_sales.drop_duplicates(inplace=True)
df_calendar.drop_duplicates(inplace=True)
df_segment.drop_duplicates(inplace=True)
df_product.drop_duplicates(inplace=True)

# Manejo de valores nulos
for df in [df_sales, df_calendar, df_segment, df_product]:
    df.fillna('No definido', inplace=True)
```

```python
print(df_sales.columns)
print(df_segment.columns)
```

```python
# 1. Primer merge entre ventas y productos
df_merged = df_sales.merge(df_product, left_on='ITEM_CODE', right_on='ITEM', how='left')

# 2. Revisar qué columnas existen después de este merge
print(df_merged.columns)
```
