```python
import pandas as pd

customer_data_file="Online Retail.xlsx"
datos= pd.read_excel(customer_data_file)
datos.head(10)
```

```python
#Eliminar todos los valores que contien Nan
datos=datos.dropna()
datos.info()
```

```python
datos_reales= datos[datos["Quantity"]>0]
datos_reales.info()
```

```python
basket = (datos_reales.groupby(["InvoiceNo","Description"]))["Quantity"].sum().unstack().reset_index().fillna(0).set_index("InvoiceNo")
basket
```

```python
#Clasificación a binario de la base previa para identificar si un articulo se compró o no
def codificar(x):
    if x<=0:
        return 0
    if x>= 1:
        return 1
basket_codificado= basket.applymap(codificar)
basket_codificado
```

```python
#Filtrado de transacciones donde se han comprado más de 1 articulo
basket_filtrado= basket_codificado[(basket_codificado >0).sum(axis=1) >=2]
basket_filtrado
```

```python
#Aplicación del algoritmo "a priori" para localizar los articulos frecuentemente comprados
from mlxtend.frequent_patterns import apriori
```

```python
#Analisis "a priori" con un soporte minimo de 3
articulos_frecuentes= apriori(basket_filtrado,min_support=0.03,use_colnames=True).sort_values("support",ascending=False).reset_index(drop=True)
articulos_frecuentes["lenght"]= articulos_frecuentes["itemsets"].apply(lambda x:len(x))
articulos_frecuentes
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(articulos_frecuentes, metric="confidence", min_threshold=0.5).sort_values("confidence",ascending=False).reset_index(drop=True)
```

```python
association_rules(articulos_frecuentes, metric="lift", min_threshold=1.01).sort_values("lift",ascending=False).reset_index(drop=True)
```
